```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('april_test_radt.csv')

#####################

april_1g = df.loc[0:359,'ground_tunshi'].reset_index(drop=True)
april_1w1 = df.loc[0:359,'wrf_tunshi_radt_30'].reset_index(drop=True)
april_1w2 = df.loc[0:359,'wrf_tunshi_radt_10'].reset_index(drop=True)
april_1w3 = df.loc[0:359,'wrf_tunshi_radt_1'].reset_index(drop=True)
april_2g = df.loc[360:719,'ground_tunshi'].reset_index(drop=True)
april_2w1 = df.loc[360:719,'wrf_tunshi_radt_30'].reset_index(drop=True)
april_2w2 = df.loc[360:719,'wrf_tunshi_radt_10'].reset_index(drop=True)
april_2w3 = df.loc[360:719,'wrf_tunshi_radt_1'].reset_index(drop=True)

import numpy as np
import matplotlib.pyplot as plt
def RE_parcent(obs,model):
    RE = ((model-obs)*100)/obs
    return RE
tunshi_radt_30_1w=RE_parcent(april_1g,april_1w1)
tunshi_radt_30_1w=np.where(tunshi_radt_30_1w>100, 100, tunshi_radt_30_1w)
tunshi_radt_10_1w=RE_parcent(april_1g,april_1w2)
tunshi_radt_10_1w=np.where(tunshi_radt_10_1w>100, 100, tunshi_radt_10_1w)
tunshi_radt_1_1w=RE_parcent(april_1g,april_1w3)
tunshi_radt_1_1w=np.where(tunshi_radt_1_1w>100, 100, tunshi_radt_1_1w)
tunshi_radt_30_2w=RE_parcent(april_2g,april_2w1)
tunshi_radt_30_2w=np.where(tunshi_radt_30_2w>100, 100, tunshi_radt_30_2w)
tunshi_radt_10_2w=RE_parcent(april_2g,april_2w2)
tunshi_radt_10_2w=np.where(tunshi_radt_10_2w>100, 100, tunshi_radt_10_2w)
tunshi_radt_1_2w=RE_parcent(april_2g,april_2w3)
tunshi_radt_1_2w=np.where(tunshi_radt_1_2w>100, 100, tunshi_radt_1_2w)
######## Hours vs GHI ########
plt.figure(figsize=(12,6),dpi=100)

plt.rcParams['axes.linewidth'] = 1 #set the value globally
plt.subplot(411)

#april_1i = df.loc[0:359,'ineichen_tunshi'].reset_index(drop=True)
#plt.plot(april_1i, 'g',label='ineichen model',linewidth=1)
plt.plot(april_1g,'b',label='ground',linewidth=1)
plt.plot(april_1w1, 'g',label='radt: 30',linewidth=1)
plt.plot(april_1w2, 'orange',label='radt: 10',linewidth=1)
plt.plot(april_1w3, 'r',label='radt: 1',linewidth=1)
#plt.title('A time series of GHI values for ineichen clear sky model, measurement and WRF-solar (radt: 1, 10 and 30 mins) in the month of april, 2018 for tunshi station')
plt.title('(e) April (1-15 days), 2018 at Tunshi station',fontsize=12)
plt.ylabel('GHI ($w/m^2$)',fontsize=10)
plt.xticks([])
plt.yticks(fontsize=12)
#legend_properties = {'weight':'bold'}
plt.legend(fontsize=8,loc='upper left')
plt.ylim(0,1400)
plt.grid()
plt.gca().legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.subplot(412)
plt.plot(tunshi_radt_30_1w,'g',linewidth=1)
plt.plot(tunshi_radt_10_1w,'orange',linewidth=1)
plt.plot(tunshi_radt_1_1w,'r',linewidth=1)
plt.ylim(-100,120)
#plt.xlabel('April 1 - April 15',fontsize=12)
plt.ylabel('Relative Error [%]',fontsize=10)
plt.yticks(fontsize=12)
#plt.legend(fontsize=6,loc='upper left')
plt.grid()
plt.title('(f)')
plt.xticks([])


##################

plt.subplot(413)

#april_2i = df.loc[360:719,'ineichen_tunshi'].reset_index(drop=True)
#plt.plot(april_2i, 'g',label='ineichen model',linewidth=1)
plt.plot(april_2g,'b',label='ground',linewidth=1)
plt.plot(april_2w1, 'g',label='radt: 30',linewidth=1)
plt.plot(april_2w2, 'orange',label='radt: 10',linewidth=1)
plt.plot(april_2w3, 'r',label='radt: 1',linewidth=1)
plt.title('(g) April (16-30 days), 2018 at Tunshi station',fontsize=12)
#plt.xlabel('Hours (UTC)',fontsize=15)
plt.ylabel('GHI ($w/m^2$)',fontsize=10)
plt.xticks([])
plt.yticks(fontsize=12)
#legend_properties = {'weight':'bold'}
#plt.legend(fontsize=8,loc='upper left')
plt.ylim(0,1400)
plt.grid()


plt.subplot(414)
plt.plot(tunshi_radt_30_2w,'g',linewidth=1)
plt.plot(tunshi_radt_10_2w,'orange',linewidth=1)
plt.plot(tunshi_radt_1_2w,'r',linewidth=1)
plt.ylim(-100,120)
#plt.xlabel('April 16 - April 30',fontsize=12)
plt.title('(h)')
plt.ylabel('Relative Error [%]',fontsize=10)
plt.xticks([])
plt.yticks(fontsize=12)
#plt.legend(fontsize=6,loc='upper left')
plt.grid()
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.1, hspace=0.3)

plt.show()

```
