```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Test2.txt',header=None, delim_whitespace=True)
df1 = df.describe()
df2=df1.T
plt.figure(figsize=(15,8),dpi=500)
plt.subplot(411)
df2.loc[:,'max'].plot(color='r',label='max')
plt.legend()
plt.ylabel('Temperature (K)',fontsize=10)
plt.title('statistics of horizontal pixels from velox filter BP08645 for the flight on 02-02-2020')
plt.subplot(412)
df2.loc[:,'min'].plot(color='g',label='min')
plt.legend()
plt.ylabel('Temperature (K)',fontsize=10)
plt.subplot(413)
df2.loc[:,'mean'].plot(color='b',label='mean')
plt.legend()
plt.ylabel('Temperature (K)',fontsize=10)
plt.subplot(414)
df2.loc[:,'std'].plot(color='m',label='std')
plt.legend()
plt.xlabel('Horizontal Pixels',fontsize=10)
plt.ylabel('Temperature (K)',fontsize=10)
plt.show()
```
