```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('df.csv')
df_tixan_oct17_22= pd.DataFrame(df.loc[3312:3455,'ground_tixan'].reset_index(drop=True))

#################14th august tunshi clear day#################
df_3km = pd.read_csv('overcast_3km_tixan.csv')
df_9km = pd.read_csv('overcast_9km_tixan.csv')
x = np.arange(0,145,1)
y_tkm = np.repeat(x, 180)
y_nkm = np.repeat(x, 60)
df_y_tkm = pd.DataFrame(y_tkm,columns=['time'])
df_y_nkm = pd.DataFrame(y_nkm,columns=['time'])
tkm = pd.concat([df_y_tkm,df_3km], axis=1)
nkm = pd.concat([df_y_nkm,df_9km], axis=1)
test3km = tkm.groupby(['time']).mean()
test9km = nkm.groupby(['time']).mean()
test_concat = pd.concat([test3km,test9km], axis=1)
df_all = pd.concat([test3km,test9km,df_tixan_oct17_22], axis=1)
df_all.reset_index(inplace=True)
df_all['time']= df_all['index']
del df_all['index']
#df_all.tail()
df_all.drop(df_all.tail(1).index,inplace=True)
```

```python
def rmse(y, y_pred):
    return np.sqrt(np.mean(np.square(y - y_pred)))
rmse_rrtmg_3=rmse(df_all.ground_tixan,df_all.overcast_rrtmg_tixan_3km)
rmse_rrtmg_9=rmse(df_all.ground_tixan,df_all.overcast_rrtmg_tixan_9km)
rmse_dhudia_3=rmse(df_all.ground_tixan,df_all.overcast_dudhia_tixan_3km)
rmse_dhudia_9=rmse(df_all.ground_tixan,df_all.overcast_dudhia_tixan_9km)
rmse_goddard_old_3=rmse(df_all.ground_tixan,df_all.overcast_goddard_old_tixan_3km)
rmse_goddard_old_9=rmse(df_all.ground_tixan,df_all.overcast_goddard_old_tixan_9km)
rmse_goddard_new_3=rmse(df_all.ground_tixan,df_all.overcast_goddard_tixan_3km)
rmse_goddard_new_9=rmse(df_all.ground_tixan,df_all.overcast_goddard_tixan_9km)

def RE_parcent(obs,model):
    RE = ((model-obs)*100)/obs
#    RE1=np.where(RE>100, 0, RE)
#    RE2=np.where(RE1<-99, 0, RE1)
    RE3=np.nan_to_num(RE)
    return RE3
RE_rrtmg_3=RE_parcent(df_all.ground_tixan,df_all.overcast_rrtmg_tixan_3km)
RE_rrtmg_9=RE_parcent(df_all.ground_tixan,df_all.overcast_rrtmg_tixan_9km)
RE_dhudia_3=RE_parcent(df_all.ground_tixan,df_all.overcast_dudhia_tixan_3km)
RE_dhudia_9=RE_parcent(df_all.ground_tixan,df_all.overcast_dudhia_tixan_9km)
RE_goddard_old_3=RE_parcent(df_all.ground_tixan,df_all.overcast_goddard_old_tixan_3km)
RE_goddard_old_9=RE_parcent(df_all.ground_tixan,df_all.overcast_goddard_old_tixan_9km)
RE_goddard_new_3=RE_parcent(df_all.ground_tixan,df_all.overcast_goddard_tixan_3km)
RE_goddard_new_9=RE_parcent(df_all.ground_tixan,df_all.overcast_goddard_tixan_9km)
```
