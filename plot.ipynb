```python
import pandas as pd
import matplotlib.pyplot as plt
df1=pd.read_csv('CTH_1km.csv',sep=' ')
df2=pd.read_csv('CTH_2km.csv',sep=' ')
plt.figure(figsize=(7,4),dpi=500)
plt.plot(df1.r_eff,df1.Btemp, 'r+', label='cloud height: 1km')
plt.plot(df2.r_eff,df2.Btemp, 'b+', label='cloud height: 2km')
plt.xlabel('effective radius ($\mu m$)', fontsize=15)
plt.ylabel('brightness temperature (k)', fontsize=15)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12, rotation=45)
plt.xlim([3,18])
plt.ylim([252.5,272.5])
plt.grid()
plt.annotate('LWC:0.1', xy=(15.3, 258.03), xytext=(16.3, 258.1),
            arrowprops=dict(arrowstyle="->"))
plt.annotate('LWC:1', xy=(15.3, 254.45), xytext=(16.3, 254.1),
            arrowprops=dict(arrowstyle="->"))
plt.annotate('LWC:0.1', xy=(15.3, 271.41), xytext=(16.3, 271.4),
            arrowprops=dict(arrowstyle="->"))
plt.annotate('LWC:1', xy=(15.3, 270.29), xytext=(16.3, 270.2),
             arrowprops=dict(arrowstyle="->"))
plt.legend()
plt.show()
```
