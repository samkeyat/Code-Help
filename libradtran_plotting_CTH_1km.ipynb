```python
%matplotlib inline
import glob
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

read_files = glob.glob(os.path.join("outfile_*.txt"))
array = []
filename= []
for files in read_files:
    data = pd.read_csv(files,header=None, sep='\,')
    array.append(data)
    filename.append(files)
    new_set = {x.replace('outfile_', '').replace('.txt', '') for x in filename}
df1 = pd.DataFrame(array)
df1.to_csv('df.csv',header=None, index=False)
df= pd.read_csv('df.csv', sep='   ', header = None)
df = df.drop(df.columns[[0, 1]], axis=1)
df = df.dropna().reset_index(drop=True)
df_bt = df.reset_index(drop=True)
df_bt.columns=['BTemp_LWC_1'] # change the parameter
df_new = pd.DataFrame(new_set, columns=['effective_radius'])
df_eff = df_new['effective_radius'].sort_values().reset_index(drop=True)
modified = pd.concat([df_eff,df_bt], axis=1)
result = modified.sort_values(['BTemp_LWC_1','effective_radius'],ascending=[1, 0]).reset_index(drop=True) #change parameter
#result
result.to_csv('LWC_1.csv', index=False,sep=' ') # change parameter
```

```python
import os
import glob
import pandas as pd
import numpy as np

allFiles = glob.glob(os.path.join("LWC_*.csv"))


np_array_list = []
for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, header=0, sep=' ')
    np_array_list.append(df.as_matrix())

comb_np_array = np.vstack(np_array_list)
big_frame = pd.DataFrame(comb_np_array)
big_frame.columns = ["r_eff","Btemp"]
#big_frame.head(10)
big_frame.to_csv('CTH_1km.csv', index=False,sep=' ')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,4),dpi=1500)
plt.plot(big_frame.r_eff,big_frame.Btemp, 'r+')
plt.xlabel('effective radius ($\mu m$)', fontsize=15)
plt.ylabel('brightness temperature (k)', fontsize=15)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12, rotation=45)
plt.xlim([3,18])
plt.ylim([269.5,272])
plt.grid()
plt.annotate('LWC:0.1', xy=(15.3, 271.41), xytext=(16.3, 271.4),
            arrowprops=dict(arrowstyle="->"))
plt.annotate('LWC:0.2', xy=(15.3, 270.83), xytext=(16.3, 270.83),
             arrowprops=dict(arrowstyle="->"))
plt.annotate('LWC:0.3', xy=(15.3, 270.62), xytext=(16.3, 270.63),
             arrowprops=dict(arrowstyle="->"))
plt.annotate('LWC:1', xy=(15.3, 270.29), xytext=(16.3, 270.2),
             arrowprops=dict(arrowstyle="->"))
plt.title('cloud height 1km', fontsize=15)
#plt.legend(fontsize=12)
plt.show()
```
