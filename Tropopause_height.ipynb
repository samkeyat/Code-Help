```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://acd-ext.gsfc.nasa.gov/Data_services/cloud_slice/data/tp_5by5.txt', skiprows=8,header=None)
df1 = df.iloc[1461:1533,:]
df1.to_csv('df1.csv', header=None,index=False)
df2 = pd.read_csv('df1.csv', delim_whitespace=True, header=None)
df2.drop(df2.columns[[0]], axis=1, inplace=True)
def TPH(p):
    eq = 44.33*(1-((p/1013.25)**0.19))
    return eq
b = TPH(df2.mean())
plt.figure(1, figsize=(10, 6))
b.plot(color= 'blue', label='10°N-15°N')
plt.xticks(fontsize=20, rotation=0)
plt.yticks(fontsize=20, rotation=0)
plt.xlabel('Months (Jan-Dec)', fontsize=20)
plt.ylabel('Altitude (km)', fontsize=20)
plt.ylim(14,17)
plt.legend(loc='upper right', fontsize=20)
#plt.savefig('TPH_10N-15N.png')
plt.show()
```
