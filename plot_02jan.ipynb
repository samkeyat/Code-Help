```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.DataFrame(np.array([[267.97748, 90.3818, 125.03123,349.42487,304.777238,296.46,309.51], [1100.371, 300.4153,56.794334,313.8531,1137.17069,271.04,290.18], [854.5057, 233.3784,49.512573,310.20544,891.305428,277.9,289.08]]),
                   columns=['ShortWave_UP', 'ShortWave_DOWN', 'LongWave_UP','LongWave_DOWN','ShortWave_UP_corrected','surface_temp','instrument_temp'])
```

```python
import datetime
from time import gmtime
from time import strftime
m= [41798,53570,67172]
time = []
for i in m:
    time.append(strftime("%H:%M:%S", gmtime(i)))
df_time = pd.DataFrame(time)
df_time.columns = ['time']
df2 = pd.concat([df_time,df],axis = 1)
df2
```

```python
fig, ax1 = plt.subplots(figsize=(14,6),dpi=500)
ax2 = ax1.twinx()
ax1.tick_params(direction='in', length=6, width=1, colors='k',
                grid_color='k', grid_alpha=0.5)
ax2.tick_params(direction='in', length=6, width=1, colors='k',
                grid_color='k', grid_alpha=0.5)
#ax2.set_xticklabels(xticklabels, fontsize=7)
ax1.xaxis.set_tick_params(labelsize=12)
ax2.xaxis.set_tick_params(labelsize=12)
ax1.yaxis.set_tick_params(labelsize=12)
ax2.yaxis.set_tick_params(labelsize=12)
ax1.plot(df2.time,df2.ShortWave_UP_corrected, 'r*', label = 'ShortWave_UP_corrected')
ax1.plot(df2.time,df2.ShortWave_DOWN, 'g*', label = 'ShortWave_DOWN')
ax1.plot(df2.time,df2.LongWave_UP, 'b*', label = 'LongWave_UP')
ax1.plot(df2.time,df2.LongWave_DOWN, 'm*', label = 'LongWave_DOWN')
ax2.plot(df2.time,df2.surface_temp, 'y*', label= 'surface_temp')
ax2.plot(df2.time,df2.instrument_temp, 'c*', label= 'instrument_temp')
ax1.set_xlabel('Time [UTC]',fontsize = 15)
ax1.set_ylabel('Irradiance ($W/M^2$)',fontsize = 15)
ax2.set_ylabel('Temperature (K)',fontsize = 15)
ax1.legend(fontsize=11, loc = 'upper left')
ax2.legend(fontsize=12, loc = 'upper right')
plt.title('Comparison of short and longwave up and down with surface and instrument temperature on 02.02.2020 (three periods)', fontsize=12)
ax2.set_ylim(265,340)
plt.show()
```
